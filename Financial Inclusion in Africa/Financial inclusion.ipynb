```python
import pandas as pd
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')
```

```python
train.info()
train.head()
train.describe()
```

Map the target data

```python
# Reload the original train DataFrame to ensure the 'bank_account' column has its original values
train = pd.read_csv('Train.csv')

# Inspect the unique values of 'bank_account' before mapping
print("Unique values in 'bank_account' before mapping:")
print(train['bank_account'].unique())

# Apply the mapping. This line will be adjusted based on the unique values found.
# Assuming 'Yes' and 'No' are the correct values, but if not, this will need modification.
train['bank_account'] = train['bank_account'].map({'Yes':1, 'No':0})

# Display info after mapping
train.info()
```

Create age bins


```python
#create the age bins
bins =[0,18,30,50,100]
labels = ['Child','Young Adult','Adult','Senior']
train['age_bins'] = pd.cut(train['age_of_respondent'], bins=bins, labels=labels, right=False)
train.info()
test['age_bins'] = pd.cut(test['age_of_respondent'], bins=bins, labels=labels, right=False)
train.drop('age_of_respondent', axis=1, inplace=True)
test.drop('age_of_respondent', axis=1, inplace=True)
train.info()
```

Transform education column


```python

train.marital_status.unique()
```

```python
family_members = ['Spouse', 'Child', 'Parent', 'Head of Household']
train['is_family'] = train['relationship_with_head'].isin(family_members).astype(int)

test['is_family'] = test['relationship_with_head'].isin(family_members).astype(int)
train.info()
train.drop('relationship_with_head', axis=1, inplace=True)
test.drop('relationship_with_head', axis=1, inplace=True)
train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['education_level'] = le.fit_transform(train['education_level'])
test['education_level'] = le.transform(test['education_level'])
train.info()
train.education_level.unique()
```

```python
train.info()
```

Marital status


```python
#has a partner
train['has_partner'] = train['marital_status'].isin(['Married/Living together']).astype(int)
test['has_partner'] = test['marital_status'].isin(['Married/Living together']).astype(int)
# 2. Ever Married vs. Never
ever_married_list = ['Married/Living together', 'Widowed', 'Divorced/Seperated']
train['ever_married'] = train['marital_status'].isin(ever_married_list).astype(int)
test['ever_married'] = test['marital_status'].isin(ever_married_list).astype(int)
# 3. Handling 'Dont know' as a specific flag
train['marital_status_unknown'] = (train['marital_status'] == 'Dont know').astype(int)
test['marital_status_unknown'] = (test['marital_status'] == 'Dont know').astype(int)
train.has_partner
```

```python
train.drop('marital_status', axis=1, inplace=True)
test.drop('marital_status', axis=1, inplace=True)
```

Define X and y



```python
X = train.drop(['uniqueid','bank_account'], axis=1)
y = train['bank_account']
X_final = pd.get_dummies(X, drop_first=True, dtype=int)
X_final.info()
```

split and train the train data


```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X_final, y, test_size=0.2,random_state=42)
```

Model 1. Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
logic = LogisticRegression(max_iter=1000,random_state=42)
logic.fit(X_train_scaled, y_train)
```

Mean Absolute Error


```python
from sklearn.metrics import mean_absolute_error
X_valid_scaled = scaler.transform(X_valid)
from sklearn.metrics import accuracy_score
y_pred = logic.predict(X_valid_scaled)
# 3. Calculate Accuracy
accuracy = accuracy_score(y_valid, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# 4. Calculate MAE (For your Zindi leaderboard)
mae = 1 - accuracy
print(f"MAE: {mae:.4f}")
# 2. Calculate MAE
mae = mean_absolute_error(y_valid, y_pred)
print(f"Mean Absolute Error: {mae}")
```

Model 2:Random Forest Classifier

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
random = RandomForestClassifier(n_estimators=300, random_state=42)
random.fit(X_train_scaled,y_train)
prediction = random.predict(X_valid_scaled)
print("Accuracy:",metrics.accuracy_score(y_valid, prediction))
```

model 3. XGBoost


```python
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=42)
xgb.fit(X_train_scaled, y_train)
xgb_pred = xgb.predict(X_valid_scaled)
xgb_accuracy = accuracy_score(y_valid, xgb_pred)
print(f"Accuracy: {xgb_accuracy:.4f}")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder # Assuming 'le' is already fitted from P2_inKC6P182

# 1. Reload the original test DataFrame to ensure all original columns are present
test = pd.read_csv('Test.csv')

# 2. Apply Label Encoding to 'education_level' using the pre-fitted 'le'
# This step was applied to 'train' before X_final was created.
test['education_level'] = le.transform(test['education_level'])

# 3. Create 'age_bins' and drop 'age_of_respondent' as per the 'train' processing
bins =[0,18,30,50,100]
labels = ['Child','Young Adult','Adult','Senior']
test['age_bins'] = pd.cut(test['age_of_respondent'], bins=bins, labels=labels, right=False)
test.drop('age_of_respondent', axis=1, inplace=True)

# 4. Drop 'uniqueid' from test as it's not a feature for the model
test_processed = test.drop('uniqueid', axis=1)

# 5. Apply one-hot encoding to the remaining categorical features to match X_train
# Based on the X_train columns, these columns were one-hot encoded.
cat_cols_for_dummies = ['country', 'location_type', 'cellphone_access', 'gender_of_respondent',
                        'relationship_with_head', 'marital_status', 'job_type']
X_test = pd.get_dummies(test_processed, columns=cat_cols_for_dummies, drop_first=True, dtype=int)

# 6. Reindex X_test to match X_train exactly
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# 7. Final Verify
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape:  {X_test.shape}")
print("Match:", list(X_train.columns) == list(X_test.columns))
```
